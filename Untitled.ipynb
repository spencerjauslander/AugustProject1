```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Homicide_df_raw = pd.read_csv("Homicides Are Falling Around the World - Homicides Are Falling Around the World.csv")
Homicide_df_raw
```

```python
new_header = Homicide_df_raw.iloc[0] 
Homicide_df = Homicide_df_raw[1:] 
Homicide_df.columns = new_header
Homicide_df
```

```python
Homicide_df_final = Homicide_df.drop(columns=["source", "URL"])
Homicide_df_final = Homicide_df_final.dropna()
Homicide_df_final = Homicide_df_final.reset_index(drop=True)
Homicide_df_final = Homicide_df_final.rename(columns={"country": "Name"})
Homicide_df_final
```

```python
Freedom_df_raw = pd.read_csv("freedom-scores.csv")
Freedom_df_raw
```

```python
Freedom_df = Freedom_df_raw.dropna(subset=['Overall Score'])
Freedom_df
```

```python
ultimate_df = pd.merge(Homicide_df_final[['Name','% change earliest to latest data', "earliest figure", "latest figure"]],Freedom_df[["Name", "Index Year", "Overall Score"]], how='left', on='Name')
ultimate_df = ultimate_df.groupby(["Name",'% change earliest to latest data', "earliest figure", "latest figure","Index Year"])
ultimate_df.first()
```
