```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Homicides Are Falling Around the World - Homicides Are Falling Around the World.csv")
df
```

```python
new_header = df.iloc[0] 
df = df[1:] 
df.columns = new_header
df
```

```python
test = df.drop(columns=["source", "URL"])
test
```

```python
test2 = test.dropna()
test2 = test2.reset_index()
test2 = test2.drop(columns=["index"])
test2
```

```python
listarray = list(np.arange(1995,2016)) 
```

```python
plt.figure(figsize=(10,10))
for i in range(len(test2["country"])):
    plt.plot(listarray,test2.iloc[i,2:23])
plt.xticks(listarray,rotation=90)
plt.xlabel("Years")
plt.ylabel("Intentional homicides (per 100,000 people)")
plt.legend(test2['country'],loc='best', fontsize = 'xx-small')
plt.show()
```
