```python
df2 = pd.read_csv("freedom-scores.csv")
df2
```

```python
df3 = df2.dropna()
df3 = df3.reset_index(drop=True)
df3
```

```python
temp = df2.loc[(df2["Name"] == "Croatia") | (df2["Name"] == "Colombia") | (df2["Name"] == "Kazakhstan") | (df2["Name"] == "Poland")]
temp = temp.loc[(temp["Index Year"] >= 1995)  & (temp["Index Year"] <= 2015)]
temp
```

```python
Croatia = temp.loc[temp["Name"] == "Croatia"]
Croatia = Croatia.sort_values("Index Year")
Croatia
```

```python
Colombia = temp.loc[temp["Name"] == "Colombia"]
Colombia = Colombia.sort_values("Index Year")
Colombia
```

```python
Kazakhstan = temp.loc[temp["Name"] == "Kazakhstan"]
Kazakhstan = Kazakhstan.sort_values("Index Year")
Kazakhstan
```

```python
Poland = temp.loc[temp["Name"] == "Poland"]
Poland = Poland.sort_values("Index Year")
Poland
```

```python
countryarray = ["Poland", "Kazakhstan", "Colombia", "Croatia"]
```

```python
polandpercent = (Poland.loc[Poland["Index Year"] == 2015]["Overall Score"].reset_index(drop=True) - Poland.loc[Poland["Index Year"] == 1995]["Overall Score"].reset_index(drop=True)) / Poland.loc[Poland["Index Year"] == 2015]["Overall Score"].reset_index(drop=True)
polandpercent = polandpercent * 100
kazakhstanpercent = (Kazakhstan.loc[Kazakhstan["Index Year"] == 2015]["Overall Score"].reset_index(drop=True) - Kazakhstan.loc[Kazakhstan["Index Year"] == 1998]["Overall Score"].reset_index(drop=True)) / Kazakhstan.loc[Kazakhstan["Index Year"] == 2015]["Overall Score"].reset_index(drop=True)
kazakhstanpercent = kazakhstanpercent * 100
colombiapercent = (Colombia.loc[Colombia["Index Year"] == 2015]["Overall Score"].reset_index(drop=True) - Colombia.loc[Colombia["Index Year"] == 1995]["Overall Score"].reset_index(drop=True)) / Colombia.loc[Colombia["Index Year"] == 2015]["Overall Score"].reset_index(drop=True)
colombiapercent = colombiapercent * 100
croatiapercent = (Croatia.loc[Croatia["Index Year"] == 2015]["Overall Score"].reset_index(drop=True) - Croatia.loc[Croatia["Index Year"] == 1996]["Overall Score"].reset_index(drop=True)) / Croatia.loc[Croatia["Index Year"] == 2015]["Overall Score"].reset_index(drop=True)
croatiapercent = croatiapercent * 100
```

```python
kazakhstanpercent
```

```python
changefreedomscore = [polandpercent.iloc[0], kazakhstanpercent.iloc[0], colombiapercent.iloc[0], croatiapercent.iloc[0]]
changefreedomscore
```

```python
colordict = {"Poland": "g",
            "Kazakhstan":"y",
            "Colombia":"r",
            "Croatia":"b"}
colordict
```

```python
combinded_df = pd.DataFrame({
    "Country Name" : countryarray,
    "% Change in Homicides: 1995 - 2015" : percenthomicides,
    '% Change in Overall Freedom Score: Earliest Date - 2015' : changefreedomscore
})
combinded_df
```

```python
combinded_df["% Change in Homicides: 1995 - 2015"] = pd.to_numeric(combinded_df["% Change in Homicides: 1995 - 2015"])
combinded_df["% Change in Homicides: 1995 - 2015"] = combinded_df["% Change in Homicides: 1995 - 2015"]*100
combinded_df
```

```python
plt.bar(combinded_df["Country Name"], combinded_df["% Change in Homicides: 1995 - 2015"], color ='g', width = .25,
        edgecolor ='grey')
plt.bar(combinded_df["Country Name"], combinded_df["% Change in Overall Freedom Score: Earliest Date - 2015"], color ='b', width = .25,
        edgecolor ='grey')
plt.ylim([-80, 40])
plt.legend(["% Change in Homicides: 1995 - 2015","% Change in Overall Freedom Score: Earliest Date - 2015"],loc='best', fontsize = 'xx-small')
plt.xlabel("Countries")
plt.ylabel("Percent Change")
plt.title("Percent change of homicides in relation to percent change of freedom scores")
plt.show()
```
